```python
import pandas as pd

dados = pd.read_csv("dados_venda.csv")
```

```python
dados.columns
```

```python
dados = dados[['oportunidade', 'localizacao', 'preco',
       'venda/alugar', 'topologia', 'estadocon', 'dist_utfpr', 'dist_praca']]
```

```python
list_preco = list()
list_dist_u = list()
list_dist_p = list()
for linha in dados.index:
    if dados.preco.iloc[linha] <= 1500:
        list_preco.append("bom")
    else:
        list_preco.append("ruim")
        
    if dados.dist_utfpr.iloc[linha] < 1:
        list_dist_u.append("perto")
    elif 1 <= dados.dist_utfpr.iloc[linha] < 2:
        list_dist_u.append("medio")
    else:
        list_dist_u.append("longe")
    
    if dados.dist_praca.iloc[linha] < 1:
        list_dist_p.append("perto")
    elif 1 <= dados.dist_praca.iloc[linha] < 2:
        list_dist_p.append("medio")
    else:
        list_dist_p.append("longe")
```

```python
dados.preco = list_preco.copy()
dados.dist_utfpr = list_dist_u.copy()
dados.dist_praca = list_dist_p.copy()
```

```python
dados_classe = pd.read_csv("dados_ori.csv")
dados["classe"] = dados_classe.classedois
```

```python
dados.oportunidade.replace({"Sem informacao":"Seminformacao", "Seminformação":"Seminformacao"}, inplace=True)
```

```python
dados.sample(40, random_state=101).to_csv("train_id3.csv", index=False)

index_test = list()
for i in dados.index:
    if i not in dados.sample(40, random_state=101).index:
        index_test.append(i)

dados.iloc[index_test].to_csv("test_id3.csv", index=False)
```
