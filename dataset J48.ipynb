```python
list_terrenos = list()
list_concervacao = list()
for bairro in dados_venda.localizacao:
    if bairro in dados_bairros.bairro.unique():
        list_terrenos.append(dados_bairros.set_index("bairro").T[bairro].values[1][0])
        list_concervacao.append(dados_bairros.set_index("bairro").T[bairro].values[2][0])
    elif bairro == 'Sub Sede':
        list_terrenos.append(dados_bairros.set_index("bairro").T["São Clemente"].values[1][0])
        list_concervacao.append(dados_bairros.set_index("bairro").T["São Clemente"].values[2][0])
    elif bairro == "São Luiz":
        list_terrenos.append(dados_bairros.set_index("bairro").T["São Luis"].values[1][0])
        list_concervacao.append(dados_bairros.set_index("bairro").T["São Luis"].values[2][0])
    elif bairro == "Jardim Universitario":
        list_terrenos.append(dados_bairros.set_index("bairro").T["Bairro Universitario"].values[1][0])
        list_concervacao.append(dados_bairros.set_index("bairro").T["Bairro Universitario"].values[2][0])
    else:
        list_terrenos.append("I")
        list_concervacao.append("R")
```

```python
dados_venda["topologia"] = list_terrenos
dados_venda["estadocon"] = list_concervacao
```

```python
dados_venda["venda"] = dados_venda["venda/alugar"].map({"venda":1, "terreno":1}).fillna(0)
dados_venda["alugar"] = dados_venda["venda"].map({0:1}).fillna(0)
```

```python
for index, item in enumerate(dados_venda["venda/alugar"].values):
    dados_venda["venda/alugar"].values[index] = item.split("/")[0]
```

```python
dados_venda["venda/alugar"] = [dado.lower() for dado in dados_venda["venda/alugar"]]
```

```python
dados_venda["preco"] = [preco.split(",")[0][3:].split("/")[0].split(" ")[-1] for preco in dados_venda["preco"]]
```

```python
precos_tratados = list()
for price in dados_venda["preco"]:
    if price[-3] == ".":
        precos_tratados.append(price[:-3])
        print(price[:-2])
    else:
        precos_tratados.append(price)
```

```python
dados_venda["preco"] = precos_tratados
dados_venda["esgoto"] = "sim"
dados_venda["pavimentacao"] = "asfalto"
```

```python
index_salvo = 0
list_valores = [str(info).strip().split(" ") for info in dados_venda["infos adicionais"]]
list_valores = [sorted(item) for item in list_valores]
for i1, info in enumerate(list_valores):  
    if len(info) == 1 and info[0] == "nan" or info[0] == "Null":
        # nan
        list_valores[i1] = ['B-0', "C-0", 'D-0', 'G-0', 'S-0']
    for i2, item in enumerate(info):
        # tratamento
        list_valores[i1][i2] = list_valores[i1][i2].replace("Q","D")
        list_valores[i1][i2] = list_valores[i1][i2].replace("o","0")
        list_valores[i1][i2] = list_valores[i1][i2].replace("a","0")
        if list_valores[i1][i2] == "M-²":
            index_salvo = i2
    try:
        if index_salvo != 0:
            del list_valores[i1][index_salvo]
    except:
        pass
                
list_valores = [" ".join(linha) for linha in list_valores]

for index, valor in enumerate(list_valores):
    
    if "B" not in valor:
        list_valores[index] += " B-0"
    if "C" not in valor:
        list_valores[index] += " C-0"
    if "D" not in valor:
        list_valores[index] += " D-0"
    if "G" not in valor:
        list_valores[index] += " G-0"
    if "S" not in valor:
        list_valores[index] += " S-0"

list_valores = [sorted(item.split()) for item in list_valores]

conj_auxiliar = pd.DataFrame(columns=["B", "C", "D", "G", "S"])
for item in list_valores:
    conj_auxiliar = pd.concat([conj_auxiliar, pd.DataFrame([
        [item[0][2], item[1][2], item[2][2], item[3][2], item[4][2]]
    ], columns=["B", "C", "D", "G", "S"])])
```

```python
for coluna in ['preco',
       'venda', 'alugar', 'residencial', 'terreno', 'comercial', 'Quitinete',
       'B', 'C', 'D', 'G', 'S', 'dist_utfpr',
       'dist_praca']:
    dados_mistos_calssificacao[coluna] = pd.to_numeric(dados_mistos_calssificacao[coluna])
```

```python
dados_mistos_calssificacao.info()
```

```python
list_classes_2 = list()
for index in dados_mistos_calssificacao.index:
    pontuacao = 0
    if dados_mistos_calssificacao.iloc[index]["oportunidade"] in ["Oportunidade", "ProntoParaMorar"]:
        pontuacao += 0.5
    if dados_mistos_calssificacao.iloc[index].alugar == 1:
        pontuacao += 1
    if dados_mistos_calssificacao.iloc[index].venda == 1:
        if dados_mistos_calssificacao.iloc[index].dist_utfpr <= 1:
            pontuacao += 0.5
    if dados_mistos_calssificacao.iloc[index].terreno == 1:
        if dados_mistos_calssificacao.iloc[index].dist_utfpr <= 1:
            pontuacao += 0.5
    if dados_mistos_calssificacao.iloc[index].Quitinete == 1:
        if dados_mistos_calssificacao.iloc[index].dist_utfpr <= 1:
            pontuacao += 2
    if dados_mistos_calssificacao.iloc[index].dist_praca <= 1:
        pontuacao += 0.1
    if  dados_mistos_calssificacao.iloc[index].B >= 1:
        pontuacao += 0.1
    if dados_mistos_calssificacao.iloc[index].C >= 1:
        pontuacao += 0.1
    if dados_mistos_calssificacao.iloc[index].D >=1:
        pontuacao += 0.1
    if dados_mistos_calssificacao.iloc[index].G >=1:
        pontuacao += 0.1
    if dados_mistos_calssificacao.iloc[index].S >= 1:
        pontuacao += 0.1
    if pontuacao >= 1:
        list_classes_2.append("sim")
    else:
        list_classes_2.append("nao")
    
```

```python
dados_mistos_calssificacao["classedois"] = list_classes_2
```

```python
dados_mistos_calssificacao.sample(40, random_state=101).to_csv("train.csv", index=False)
```
